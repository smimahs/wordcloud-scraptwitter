```python
from twitter_scraper import get_tweets
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
from wordcloud import WordCloud
```

```python
file = open('output.txt','w')
t=[]
for tweet in get_tweets('username'):
    t.append(tweet['text'])
    file.write(tweet['text'])

file.close()
```

```python
word = open('output.txt', 'r',encoding='utf-8').read()
```

```python
out = WordCloud(
    background_color='white',
    max_words=2000
)

out.generate(word)
```

```python
mask = np.array(Image.open('alice_mask.png'))    
```

```python
out = WordCloud(background_color='white', max_words=2000, mask=mask)

# generate the word cloud
out.generate(word)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(out, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
mask = np.array(Image.open('alice_mask.png'))    
```

```python
from wordcloud_fa import WordCloudFa
```

```python
file = open('output-fa.txt','w')
t=[]
for tweet in get_tweets('username'):
    t.append(tweet['text'])
    file.write(tweet['text'])

file.close()
```

```python
word = open('output-fa.txt', 'r').read()
```

```python
out = WordCloudFa(persian_normalize=True,collocations=False,background_color='white', max_words=2000, mask=mask)

# generate the word cloud
out.generate(word)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(out, interpolation='bilinear')
plt.axis('off')
plt.show()
```
